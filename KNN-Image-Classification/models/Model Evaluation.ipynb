# Model Evaluation

###  Importing Required Libraries

```python
import pandas as pd 
import numpy as np
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
from sklearn.metrics import classification_report, confusion_matrix
```

## 1) Loading test data and trained model

```python
data = datasets.load_iris()
x_test = np.load("test_data.npy")
y_test = np.load("test_labels.npy")
classifier = joblib.load("knn_model.joblib")
```

## 2) Evaluating the Model

```python
y_pred = classifier.predict(x_test)
accuracy = classifier.score(x_test, y_test)*100
```

```python
print(f'Accuracy achieved is  {accuracy:.2f}%\n')
report = classification_report(y_test ,y_pred, target_names= data.target_names)
print(report)
```

```python
cf = confusion_matrix(y_test, y_pred)
print(cf)
```

## 3) Visualising the Confusion Matrix

```python
plt.figure(figsize=(5, 3))
sns.heatmap(cf, annot=True, fmt='g', xticklabels= data.target_names, yticklabels= data.target_names)
plt.title("Confusion Matrix")
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
